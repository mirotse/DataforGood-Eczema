```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2, perplexity = 5)
trans_tsne = tsne.fit_transform(df_trans)
```

```python
df_trans_tsne = pd.DataFrame(trans_tsne)
df_trans_tsne.index = df_trans.index
df_trans_tsne = df_trans_tsne.join(df.set_index(['sample_id']))
df_trans_tsne.rename(columns={0:'tsne_one', 1:'tsne_two', 2:'iris_three'}, inplace=True)
```
